## Section 4.1: Is the first to get drafted the best?
In the NBA, there is a common stereotype that the number one draft pick of a year will always have the best rookie season. In today's basketball talks, `Victor Wembenyama`, this year's number 1 pick, is a very hot topic and everyone believes that he will go on to win the best rookie award. However, I wanted to justify this stereotype myself using the NBA dataset. This got me interested in the `correlation` between the `draft number` of a player and their `points` scored. This is becuase the number of points per game scored is often used as a measure to see how a player is performing. While this may not be comprehensive of the full picture, it provides a good enough picture. In the following code cell, I narrow down the data to the columns that I need.

```python
relevant_columns = ['pts','draft_number']
```

Since I am checking the performance of the players drafted in a specific year during their rookie season, it is necessary to use a complex conditional in order to narrow down to a specific year and season. Ideally, I would perform regression on every year and season in the dataset however in this project, I will only perform it for the most recent year. I run the code to find the complex conditional in the following code cell.

```python
df_2 = df[(df['draft_year']=='2022')&(df['season']=='2022-23')]
df_relevant = df_2[relevant_columns]
```

In the following code cell, I had to use the command `astype()`. This is because the dtype of my column `draft_number` was initially set to be `object`. Due to this, I was unable to use this column as an integer as it is supposed to be used. Therefore, I did some research and found a command in order to change the dtype of this column.

```python
df_relevant['draft_number']=df_relevant['draft_number'].astype('float64')
df_relevant
```

In the following code cell, I checked the actual correlation between the points and draft number of a player drafted in 2022 and found that it is as expected, a high negative correlation between the two. This makes the data suitable for performing regression.

```python
df_relevant.corr()
```

## Section 4.2: Visualization of the data
Next, I plotted a scatter graph for to better demonstrate the correlation between these variables visually and to understand how the points scored vary as the draft number increases. We find that the trend is just as expected with the points scored decreasing on average as draft number increases.

```python
import matplotlib.pyplot as plt
plt.scatter(df_relevant['draft_number'],df_relevant['pts'])
plt.xlim(0, 61)
plt.ylim(0, 35)
plt.show()
```

## Section 4.3: Finding the regression line
The next step was calculating the **slope** using the formula:
> $$\text{slope} = r * \frac{SD(y)}{SD(x)}$$

Where $r$ is the correlation coefficient between $X$ and $y$, and $SD$ stands for standard deviation. 

```python
slope = df_relevant.corr()['pts']['draft_number']*(df_relevant['pts'].std()/df_relevant['draft_number'].std())
slope
```

Then I calculated the **y-intercept** of the regression line using the formula as follows:

> $$\text{intercept} = \text{avg}(y) - \text{slope} * \text{avg}(X)$$

```python
intercept = df_relevant['pts'].mean() - slope *df_relevant['draft_number'].mean()
intercept
```

## Section 4.4: What does the regression line look like?
Now that I have found the equation to the regression line, lets see how it would look on the scattergraph as a best fit. In the following cell, I imported the `seaborn` library in order to visualize the regression line that I derived in the last two steps.

```python
import seaborn as sns

sns.regplot(data=df_relevant, x="draft_number", y="pts", ci=None, line_kws={'color':'red'})
```

## Section 4.5: Prediction and residual
Using the regression model that we have generated, I will create two additional columns in the dataframe, the prediction column and the residual. The residual will give us a measure of how close the prediction is to the actual value.

```python
df_relevant['pts_prediction'] = slope*df_relevant['draft_number']+intercept
df_relevant
```

```python
df_relevant['residual'] = df_relevant['pts']-df_relevant['pts_prediction']
df_relevant
```
